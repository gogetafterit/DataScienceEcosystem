```python

from github import Github, Auth
from google.colab import drive
import os

# 1️⃣ Montar Google Drive
drive.mount('/content/drive', force_remount=True)

# 2️⃣ Configuración de GitHub
# Nunca pongas tu token directamente en el notebook
# [REMOVED SECRET]token = os.environ.get("GITHUB_TOKEN")
repo_name = "gogetafterit/DataScienceEcosystem"  # usuario/repositorio
branch_name = "main"  # O "master"

# 3️⃣ Ruta del archivo en Drive
file_path = "/content/drive/MyDrive/Colab Notebooks/DataScienceEcosystem.ipynb"
if not os.path.exists(file_path):
    raise FileNotFoundError(f"No se encontró el archivo: {file_path}")

commit_message = "Add or update datascience ecosystem notebook"

# 4️⃣ Autenticación en GitHub
g = Github(auth=Auth.Token(token))
repo = g.get_repo(repo_name)

# 5️⃣ Leer contenido del archivo
with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()

# 6️⃣ Subir o actualizar el archivo
file_name_in_repo = "datascience_ecosystem.ipynb"  # Nombre en GitHub

try:
    # Intentar obtener el archivo en el repo
    existing_file = repo.get_contents(file_name_in_repo, ref=branch_name)
    # Si existe, actualizarlo
    repo.update_file(
        path=existing_file.path,
        message=commit_message,
        content=content,
        sha=existing_file.sha,
        branch=branch_name
    )
    print(f"✅ File '{file_name_in_repo}' updated successfully in GitHub!")

except Exception as e:
    # Si no existe, crearlo
    repo.create_file(
        path=file_name_in_repo,
        message=commit_message,
        content=content,
        branch=branch_name
    )
    print(f"✅ File '{file_name_in_repo}' created successfully in GitHub!")
```
