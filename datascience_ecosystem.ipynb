```python

from github import Github, Auth
from google.colab import drive
import os
import json

drive.mount('/content/drive', force_remount=True)

# [REMOVED SECRET]token = os.environ.get("GITHUB_TOKEN")
repo_name = "gogetafterit/DataScienceEcosystem"
branch_name = "main"

file_path = "/content/drive/MyDrive/Colab Notebooks/DataScienceEcosystem.ipynb"
if not os.path.exists(file_path):
    raise FileNotFoundError(f"No se encontr√≥ el archivo: {file_path}")

commit_message = "Add or update datascience ecosystem notebook"

with open(file_path, "r", encoding="utf-8") as f:
    notebook = json.load(f)

# Buscar y reemplazar cualquier token en las celdas de c√≥digo
for cell in notebook.get("cells", []):
    if cell.get("cell_type") == "code":
        cell["source"] = [
# [REMOVED SECRET]            for line in cell.get("source", [])
        ]

# Convertir de nuevo a JSON para subirlo
content = json.dumps(notebook, ensure_ascii=False)

g = Github(auth=Auth.Token(token))
repo = g.get_repo(repo_name)

file_name_in_repo = "datascience_ecosystem.ipynb"  # Nombre en GitHub

try:
    # Intentar obtener el archivo en el repo
    existing_file = repo.get_contents(file_name_in_repo, ref=branch_name)
    # Si existe, actualizarlo
    repo.update_file(
        path=existing_file.path,
        message=commit_message,
        content=content,
        sha=existing_file.sha,
        branch=branch_name
    )
    print(f" File '{file_name_in_repo}' updated successfully in GitHub!")

except Exception as e:
    # Si no existe, crearlo
    repo.create_file(
        path=file_name_in_repo,
        message=commit_message,
        content=content,
        branch=branch_name
    )
    print(f" File '{file_name_in_repo}' created successfully in GitHub!")
```
